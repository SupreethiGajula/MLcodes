```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
# Generate sample dataset
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x) + np.random.normal(0, 0.2, 100)
# Apply LOWESS
lowess = sm.nonparametric.lowess(y, x)
# Extract the smoothed values
x_smoothed = lowess[:, 0]
y_smoothed = lowess[:, 1]
# Plot the original data and the LOWESS curve
plt.scatter(x, y, label='Original Data')
plt.plot(x_smoothed, y_smoothed, color='red', label='LOWESS')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.title('Locally Weighted Scatterplot Smoothing (LOWESS)')
plt.show()
```
